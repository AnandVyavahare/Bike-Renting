## Outlier Analysis 

```python
plt.figure(figsize=(20,16))
for i,col in enumerate(list(bike_df.columns)[0:13]):
    plt.subplot(4,5,i+1)
    plt.boxplot(bike_df[col])
    plt.title(col)
    plt.tight_layout()
    #plt.legend()
```

Defining a function using Inter Quartile Range to detect the outliers

```python
def outliers_iqr_method(df):
    outliers = np.array([])
    for col in df.columns:
        q1, q3 = np.percentile(df[col],[25,75])
        iqr = q3 - q1
        lower_limit = q1 - 1.5*iqr
        upper_limit = q3 + 1.5*iqr
        indexes = df[(df[col] < lower_limit) | (df[col] > upper_limit)].index
        outliers = np.append(outliers,indexes)
    print(indexes)
    return outliers
        
        
```

```python
outliers = np.unique(outliers_iqr_method(bike_df))
```

```python
bike_df.drop(outliers, inplace = True)
```

# Modeling

```python
#Standardizing the data
scaler = StandardScaler()
scaled_df = scaler.fit_transform(bike_df.iloc[:,:12])
print('Shape of scaled data:',scaled_df.shape)
```

```python
#Train Test Split:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(scaled_df, bike_df.iloc[:,-1], test_size = 0.2, random_state = 42)

X_train.shape, X_test.shape ,y_train.shape, y_test.shape
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
#Fitting the model
lr.fit(X_train,y_train)
```

```python
#Predicting on test data
y_pred_lr = lr.predict(X_test)
```

### Performance of Linear Regression Model

```python
#Visualzing Predicted vs Actual Values
sns.scatterplot(y_test,y_pred_lr)
plt.xlabel("Actual Values")
plt.ylabel('Predicted Values')
plt.title('Predicted vs Actual Values')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('Coefficients: ' ,lr.coef_)
print('Y intercept:%.3f' %lr.intercept_)
print('Mean Absolute Error: %.2f' %mean_absolute_error(y_test, y_pred_lr))
print('Mean Squared Error: %.2f' %mean_squared_error(y_test, y_pred_lr))
print('Root Mean Squared Error: %.2f' %np.sqrt(mean_squared_error(y_test, y_pred_lr)))
print('Coefficient of Determination(R^2): %.3f' %r2_score(y_test, y_pred_lr))
print('Adjusted R squared(R^2 Adjusted): %.3f' 
      %(1-(1-r2_score(y_test, y_pred_lr))*((len(X_test)-1)/(len(X_test)-len(X_test[0])-1))))
```

## Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
```

```python
#Parameters for Hyperparamter Tuning
parameter_grid = {'n_estimators':[90,100,110],'max_depth':[20,25,30],'min_samples_split':[30,40,50],'min_samples_leaf':[3,5,7],
                  'criterion':['mse','mae']}
print(parameter_grid)
```

```python
#Using GridSearchCV for hyperparameter tuning
from sklearn.model_selection import GridSearchCV
rf_grid = GridSearchCV(estimator = rf, param_grid = parameter_grid, cv = 3, verbose = 3, n_jobs = -1)
```

```python
#Fitting the model
rf_grid.fit(X_train,y_train)
```

```python
#Checking the best parameters predicted by our model
rf_grid.best_params_
```

```python
#Predicting on test data
y_pred_rf = rf_grid.predict(X_test)
```

### Performance of Random Forest Regressor Model

```python
#Evaluation Metrics
print('Mean Absolute Error: %.2f' %mean_absolute_error(y_test, y_pred_rf))
print('Mean Squared Error: %.2f' %mean_squared_error(y_test, y_pred_rf))
print('Root Mean Squared Error: %.2f' %np.sqrt(mean_squared_error(y_test, y_pred_rf)))
```

## KNN 

```python
from sklearn.neighbors import KNeighborsRegressor
```

### Decinding Value of k

```python
from sklearn.model_selection import cross_val_score
```

**Deciding value of K in range of 1 to 50 with accuracy rate and error rate using Cross Validation**

```python
accuracy_rate = []
for i in range(1,50):
    knn = KNeighborsRegressor(i)
    score = cross_val_score(knn,scaled_df, bike_df.iloc[:,-1], cv=10)
    accuracy_rate.append(score.mean())
```

```python
error_rate = []
for i in range(1,50):
    knn = KNeighborsRegressor(i)
    score = cross_val_score(knn,scaled_df, bike_df.iloc[:,-1], cv=10)
    error_rate.append(1-score.mean())
```

```python
#Plotting score vs K value
plt.figure(figsize=(10,8))
#plt.plot(range(1,50),accuracy_rate,linestyle ='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.plot(range(1,50),error_rate,linestyle ='dashed',marker='o',markerfacecolor='blue',markersize=10)
plt.title('Error Rate vs K value')
#plt.title('Accuracy Rate vs K value')
plt.xlabel('Value of K')
plt.ylabel('Error Rate')
#plt.ylabel('Accuracy Rate')
```

**Here we can see that after k=7 there is only decrease in the performance, hence we choose 7 as value of k**
